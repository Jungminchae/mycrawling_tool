```python
from spyders.Crawling import mycrawling_tool
import selenium
from bs4 import BeautifulSoup
import re
import numpy as np
import time
from selenium.webdriver.common.keys import Keys
import pickle
```

```python
def naver_page_open():
    browser = mc.browser_open_with_url('D:/jupyter/chrome_driver/chromedriver.exe', 'https://datalab.naver.com/shoppingInsight/sCategory.naver')
    browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[1]/span/label[3]').click()
    browser.find_element_by_xpath('//*[@id="18_device_0"]').click()
    browser.find_element_by_xpath('//*[@id="19_gender_1"]').click()
    browser.find_element_by_xpath('//*[@id="20_age_2"]').click()
    time.sleep(1)
    return browser
```

```python
def key_words_crawling(browser):
    body = browser.page_source
    body = BeautifulSoup(body, 'html.parser')
    ranks = body.find_all('a', attrs={'class':'link_text'})
    item_ranks = [re.sub('[0-9]',"",item.text.strip()) for item in ranks]
    number = np.arange(1,21).astype(str)
    return number, item_ranks
```

```python
def category_change(browser ,cat = ''):
    browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/span').click()
    time.sleep(1)
    if cat == '패션잡화':
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/ul/li[2]/a').click()
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/a/span').click()
    
    elif cat =='화장품/미용':
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/ul/li[3]/a').click()
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/a/span').click()
    
    elif cat =='디지털/가전':
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/ul/li[4]/a').click()
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/a/span').click()
    
    elif cat =='가구/인테리어':
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/ul/li[5]/a').click()
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/a/span').click()
    
    elif cat =='출산/육아':
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/ul/li[6]/a').click()
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/a/span').click()    
        
    elif cat =='식품':
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/ul/li[7]/a').click()
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/a/span').click()
        
    elif cat =='스포츠/레저':
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/ul/li[8]/a').click()
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/a/span').click()

    elif cat =='생활/건강':
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/ul/li[9]/a').click()
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/a/span').click()
        
    elif cat =='여가/생활편의':
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/ul/li[10]/a').click()
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/a/span').click()
        
    elif cat =='면세점':
        # 임시로 수동 스크롤 다운
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/ul/li[11]/a').click()
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/a/span').click()
```

```python
def make_dictionary(number, item_ranks, total, cat):
    temp =dict()
    for i,j in zip(number, item_ranks):
        temp[i] = j
    total[cat] = temp
    return total
```

```python
category_list = ['패션의류', '패션잡화' ,'화장품/미용', '디지털/가전', '가구/인테리어', '출산/육아', '식품', '스포츠/레저', '생활/건강','여가/생활편의']
total = dict()
```

```python
# defalut 패션의류
mc = mycrawling_tool.mycrawler()
browser = naver_page_open()
idx, ranks = key_words_crawling(browser)
total = make_dictionary(idx, ranks, total, category_list[0])
time.sleep(1)
```
